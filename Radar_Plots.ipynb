```python
import pandas as pd
import math
AA=pd.DataFrame(columns=['criteria','weight','beat','buy','scoreBeat','scoreBuy'])
print(AA)
```

```python
AA['criteria']=['risk','economicImpact','easeImplimentation']
print(AA)
```

```python
AA['weight']=[0.4,0.3,0.3]
print(AA)
```

```python
AA['beat']=[8,6,9]
AA['buy']=[8,8,4]
for index, row in AA.iterrows():
  AA['scoreBeat'][index]=row['beat']*row['weight']
  AA['scoreBuy'][index]=row['buy']*row['weight']
print(AA)
```

```python
totalScoreBeat=0
totalScoreBuy=0
for index, row in AA.iterrows():
  totalScoreBeat+=row['scoreBeat']
  totalScoreBuy+=row['scoreBuy']
print('the total score for Beat is {:.2f} and for Buy is{:.2f}'.format(totalScoreBeat,totalScoreBuy))
```

```python
import numpy as np
import matplotlib.pyplot as plt
axisLocations=np.linspace(start=0, stop=2*np.pi, num=len(AA['beat']), endpoint=False)
axisLocations=np.concatenate((axisLocations,[axisLocations[0]]))
ratingsA=AA['beat']
ratingsB=AA['buy']
ratingsACircular=np.concatenate((ratingsA,[ratingsA[0]]))
ratingsBCircular=np.concatenate((ratingsB,[ratingsB[0]]))

plt.figure(figsize=(8,8))
plt.subplot(polar=True)
plt.plot(axisLocations, ratingsACircular, 'o-', linewidth=2, label='Beat')
plt.plot(axisLocations, ratingsBCircular, 'D-', linewidth=2, label='Buy')

plt.title('Project Options Comparision', size=20)
line, labels=plt.thetagrids(np.degrees(axisLocations), labels=AA['criteria'])
plt.legend(loc='lower center', borderaxespad=-10)
plt.show()
```
